## Book Class
This class creates a book with a title, it's author, the isbn, if it's available, and the due date to return it. The books can be checked out, returned, and they can be overdue.

```python
import datetime

class Book:

  def __init__(self, title, author, isbn, available = True, due_date = None):
        # Sets all 5 attributes
        self.title = title
        self.author = author
        self.isbn = isbn
        self.available = available
        self.due_date = due_date
        pass

  def checkout(self, days_valid = 21):
    """
    Checks if the book is available, if not it prints a message. If it's
    available it changes the availability, adds a due date, and prints a
    message.

    Args:
        days_valid: number of days the book can be checked out without
          being overdue. The default is 21 days.
    """
    if self.available == True:
      self.available = False
      self.due_date = datetime.date.today() + datetime.timedelta(days=days_valid) # Today + Days is good for
      print(f"Checked out '{self.title}'. Due date: {self.due_date}")
    else:
      print(f"The book '{self.title}' is not available.")
    pass

  def return_book(self):
    """Makes the book available and deletes the due date. Prints
    a message"""
    self.available = True
    self.due_date = "Not Checked Out"
    print(f"The book '{self.title}' was returned.")
    pass

  def is_overdue(self):
    """Returns true if the book is overdue (today is a later day than due date)"""
    if not self.available:
            return datetime.date.today() > self.due_date # True if today is a later day than the due date
    return False

  def __str__(self):
    """
    Formats:
    'Title' by Author | ISBN: ISBN | Available
    'Title' by Author | ISBN: ISBN | Checked Out (Due: Date)
    """
    status = "Available" if self.available else f"Checked Out (Due: {self.due_date})"
    return f"'{self.title}' by {self.author} | ISBN: {self.isbn} | {status}"
```

## Other Methods

In this section the methods to search for books and to make an overdue report of the books are defined.


```python
#Searches the library
def search_library(query, library):
  """
  Searches in the library by author, title, and isbn. Returns a
  printed message.

  Args:
    query: the argument the uer inputs to check against.
    library: the dictionary with the books.
  """
  print(f"\n--- Search Results for '{query}' ---")
  found = False
  for book in library.values():
      if query.lower() in book.title.lower() or query.lower() in book.author.lower() or query.lower() in book.isbn.lower():
        print(book)
        found = True
  if not found:
        print("No matching books found.")

# Overdue report function
def overdue_report(library):
  """
  Prints a formated report about the overdue books.

  Args:
    library: the dictionary with the books.
  """
  print("\n--- Overdue Report ---")
  overdue_found = False
  for book in library.values():
    if book.is_overdue():
      print(book)
      overdue_found = True
  if not overdue_found:
        print("No overdue books.")
```
